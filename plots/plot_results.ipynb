```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# training without data augmentation
# for 200 epochs
results = pd.read_csv('log_features_3fc_200epochs.csv', sep =',')
```

```python
results.head()
```

```python
results.plot(x = 'epoch', y = ['acc', 'val_acc'], kind = 'line')
plt.show()
```

```python
results.plot(x = 'epoch', y = ['loss', 'val_loss'], kind = 'line')
plt.show()
```
